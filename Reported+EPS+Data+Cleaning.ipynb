# EPS Data Cleaning
The earnings data scraped from the MarketWatch website contained many 'NaN' values due to a companies having 5-year data that ranged from 2005-2017. Given that the earnings dates data ranged from 2012-Present, the data from 2005-2011 was deleted from the table.

```python
import pandas as pd
reportedEPS = pd.read_excel('/Users/Bert/Desktop/ReportedEPS.xlsx')
```

```python
reportedEPS.head(10) # reported EPS data table 
```

```python
reportedEPS.T.head(10) # transposed table for easier handling
```

```python
new_report = reportedEPS.fillna(0).T # perform fill of Nan values and transpose table
```

```python
for i in range(2005,2012): del new_report[i] # delete years for which earnings dates are unavailable
```
